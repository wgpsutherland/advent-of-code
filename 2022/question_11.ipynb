```python
from utils import get_puzzle_input_as_rows
import re
```

```python
def question_11_a():

    monkey_num_regex = re.compile(r"Monkey (\d):")
    items_regex = re.compile(r"Starting items: (.+)")
    operation_regex = re.compile(r"Operation: new = old (.+) (.+)")
    test_div_regex = re.compile(r"Test: divisible by (\d+)")
    if_true_regex = re.compile(r"If true: throw to monkey (\d)")
    if_false_regex = re.compile(r"If false: throw to monkey (\d)")

    monkeys = {}

    monkey_num = -1

    rows = get_puzzle_input_as_rows(2022, 11, test=False)
    for row in rows:

        if (match := monkey_num_regex.match(row)):
            monkey_num = int(match.group(1))
            monkeys[monkey_num] = {
                "items": None,
                "op_operator": None,
                "op_value_str": None,
                "test": None,
                "true_monkey": None,
                "false_monkey": None,
                "inspections": 0
            }
        elif (match := items_regex.match(row)):
            monkeys[monkey_num]["items"] = [int(x) for x in match.group(1).split(", ")]
        elif (match := operation_regex.match(row)):
            monkeys[monkey_num]["op_operator"] = match.group(1)
            monkeys[monkey_num]["op_value_str"] = match.group(2)
        elif (match := test_div_regex.match(row)):
            monkeys[monkey_num]["test"] = int(match.group(1))
        elif (match := if_true_regex.match(row)):
            monkeys[monkey_num]["true_monkey"] = int(match.group(1))
        elif (match := if_false_regex.match(row)):
            monkeys[monkey_num]["false_monkey"] = int(match.group(1))
            
    for round in range(0, 20):

        print(f"round {round}")
        
        for monkey_num in range(len(monkeys)):

            print(f"monkey {monkey_num}")

            monkey = monkeys[monkey_num]

            while len(monkey["items"]) > 0:

                monkey["inspections"] += 1

                item = monkey["items"].pop(0)  # removes from the front of the list

                print(f"insepcting item of worry value {item}")

                # inspection
                op_value = item if monkey["op_value_str"] == "old" else int(monkey["op_value_str"])
                if monkey["op_operator"] == "*":
                    item = item * op_value
                else:
                    item = item + op_value

                print(f"worry multiplied to {item}")
                
                item = item // 3

                print(f"monkey gets bored, value becomes {item}")


                if item % monkey["test"] == 0:
                    monkeys[monkey["true_monkey"]]["items"].append(item)
                    print(f"thrown to monkey {monkey['true_monkey']}")
                else:
                    monkeys[monkey["false_monkey"]]["items"].append(item)
                    print(f"thrown to monkey {monkey['false_monkey']}")

                print()
        
        print()

        for key, monkey in monkeys.items():
            print(key, monkey)
    
        print()

        inspections = sorted([monkey["inspections"] for monkey in monkeys.values()], reverse=True)

        print(inspections[0] * inspections[1])

question_11_a()  # 76728
```
